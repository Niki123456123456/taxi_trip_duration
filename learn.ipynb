```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

weather = pd.read_csv("train_weather.csv")


weather.info()
```

```python
print(weather.columns.tolist())
```

```python
features = [
    'vendor_id', 'pickup_longitude', 'pickup_latitude',
    'passenger_count', 
    'pickup_hour', 'pickup_day', 
]

features2 = [
    'vendor_id', 'pickup_longitude', 'pickup_latitude',
    'dropoff_longitude', 'dropoff_latitude',
    'passenger_count', 
    'pickup_hour', 'pickup_day'
]

features_weather = [
    'vendor_id', 'pickup_longitude', 'pickup_latitude',
    'passenger_count', 
    'pickup_hour', 'pickup_day',  'temp_C'
]

x = weather[features]
x2 = weather[features2]
x_weather = weather[features_weather]

y = np.log1p(weather['trip_duration'])

x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size=0.2, random_state=42)
x_train2, x_valid2, _, _ = train_test_split(x2, y, test_size=0.2, random_state=42)
x_train_weather, x_valid_weather, _, _ = train_test_split(x_weather, y, test_size=0.2, random_state=42)

model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=300, learning_rate=0.1, max_depth=6, random_state=42)
model2 = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=300, learning_rate=0.1, max_depth=6, random_state=42)
model_weather = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=300, learning_rate=0.1, max_depth=6, random_state=42)

model.fit(x_train, y_train)
model2.fit(x_train2, y_train)
model_weather.fit(x_train_weather, y_train)


```

```python
# x_valid = weather[features]
# x_valid_weather = weather[features_weather]

pred = model.predict(x_valid)
pred2 = model2.predict(x_valid2)
pred_weather = model_weather.predict(x_valid_weather) 
```

```python
from sklearn.metrics import mean_squared_error

rmsle = mean_squared_error(y_valid, pred, squared=False)
rmsle2 = mean_squared_error(y_valid, pred2, squared=False)
rmsle_weather = mean_squared_error(y_valid, pred_weather, squared=False)

print(f"RMSLE: {rmsle:.4f}")
print(f"RMSLE2: {rmsle2:.4f}")
print(f"RMSLE with weather: {rmsle_weather:.4f}")
```
